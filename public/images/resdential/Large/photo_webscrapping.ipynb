```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
driver.get('https://dreamdevelopersnepal.com/gallery/')
```

```python
img=[]
name=[]
for i in range(1,31):
    img_path=driver.find_element(By.XPATH,f'/html/body/div[2]/div[4]/div/div/div/main/article/div/div/div/form/div/div/div[{i}]/a').get_attribute('href')
    img.append(img_path)
    # namesir=driver.find_element(By.XPATH,f'/html/body/div[2]/div[4]/div/div/div/main/article/div/div/div/div[{i}]/div/div[1]/div/p/a[2]').get_attribute('href')
    # name.append(namesir)
   
```

```python
import requests
import os

path = r"C:\Users\Ryzen\OneDrive\Desktop\website\images\gallery"

os.makedirs(path, exist_ok=True)

for image_url, name_url in zip(img, name):

    place = name_url.split('/')[-2].split('-')[-1].upper()
    filename = f"{place}.jpg"

    r = requests.get(image_url)

    with open(f"{path}/{filename}", "wb") as f:
        f.write(r.content)
```
